# Visualizing data


### HeatMap


```python
plt.figure(dpi=100)
corr = df.corr()# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True,
            cmap='viridis')
```

### Summary
#### Sales and Profit have direct relation
#### Quantity and Profit aslo have a small but direct relation
#### Discount and Profit have inverse relation
#### Sales and Discount have inverse relation



### Scatterplot


```python
plt.figure(dpi=100)
plt.title("Profit vs Sales")
sns.scatterplot(data=df,x='Profit',y='Sales',alpha=0.5)
```

```python
plt.figure(dpi=100)
plt.title("Profit vs Discount")
sns.scatterplot(data=df,x='Discount',y='Profit',alpha=0.5)
```

```python
plt.figure(dpi=100)
x1 = df['State'].value_counts()
x1 = x1.head(10)
x1.plot(kind = "bar",figsize=(10,6))
plt.ylabel("Numbers")
plt.xlabel("State")
plt.title("States with Top Sales")
```

```python
plt.figure(dpi=100)
x1 = df['City'].value_counts()
x1 = x1.head(10)
x1.plot(kind = "bar",figsize=(10,6))
plt.ylabel("Numbers")
plt.xlabel("Cities")
plt.title("Cities with Top Sales")
```
